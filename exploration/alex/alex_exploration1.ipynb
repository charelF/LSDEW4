```python
columns = df_20210101_auto.columns

print(df_20210101_auto.isna().any())

tit = df_20210101_auto
```

```python
df_20210101_auto = df_20210101_auto.sort_values(by="title").reset_index(drop=True)
```

```python

def column_merger(columns):
    columns = ['%s%s' % (a, '_%s' % b if b else '') for a, b in columns]
    return columns

def count_unique(series):
    return len(np.unique(series))

def mean_absolute_deviation(x):
    return np.mean(np.abs(x - np.mean(x)))
```

```python
simple_stats = {
    "total": [np.mean, np.sum, kurtosis]
}


df_temp = df_20210101_auto
df_temp = df_temp.astype(dtype= {"total":"float64", "hour": "int64"})
df_temp = df_temp.groupby(by=["hour"]).agg(simple_stats).reset_index()
df_temp.columns = column_merger(df_temp.columns)

df_temp
```
